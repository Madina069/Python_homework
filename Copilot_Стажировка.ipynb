```python
from transformers import AutoTokenizer, AutoModelForCausalLM, BitsAndBytesConfig
import re

quantization_config = BitsAndBytesConfig(load_in_4bit=True)

tokenizer = AutoTokenizer.from_pretrained("google/gemma-2-9b-it")
model = AutoModelForCausalLM.from_pretrained(
    "google/gemma-2-9b-it",
    quantization_config=quantization_config)

chat = [
    { "role": "user", "content": "What is your name?" },
]
prompt = tokenizer.apply_chat_template(chat, tokenize=False, add_generation_prompt=True)

inputs = tokenizer.encode(prompt, add_special_tokens=False, return_tensors="pt")
outputs = model.generate(input_ids=inputs.to(model.device), max_new_tokens=150)
output_decode = tokenizer.decode(outputs[0])
print(output_decode)

pattern = r'<start_of_turn>model\s*(.*?)\s*<end_of_turn>'
match = re.search(pattern, output_decode, re.DOTALL)

if match:
    answer = match.group(1)
    print("Extracted answer:", answer)
else:
    print("No match found")
```

# **Preparation of the Datasets**

```python
import pandas as pd
import numpy as np
import copy
```

```python
#raw_queries_df = pd.read_csv ('/content/raw_queries (3).csv', sep = ';')
```

```python
raw_queries_df = pd.read_csv('raw_queries (3).csv', sep = ';')
```

```python
variables = {
    "queries": 10, #количество Queries
    "samples_per_query": 10, #количество Samples на один Query
}

lines = variables["queries"] * variables["samples_per_query"]
#количество строк, взятых из исходного датасета для создания тестовой выборки
```

```python
def dataset_test_sampling(dataset):
    '''Функция для создания тестовой выборки с определенным количеством строк из исходного датасета'''

    test_df = dataset.sample(n=lines, random_state=15, replace=False)

    return test_df
```

```python
def df_for_model(queries_df, test_df, dataset_name):
    dataset = dataset_name
    queries = queries_df[queries_df['dataset'] == dataset]
    queries_df = pd.DataFrame(np.repeat(queries, variables["samples_per_query"], axis=0), columns=queries.columns)
    model_df = copy.deepcopy(test_df)
    model_df["raw query"] = queries_df["raw query"].values

    return model_df
```

**SST2**

```python
df_SST2 = pd.read_parquet("hf://datasets/nyu-mll/glue/sst2/train-00000-of-00001.parquet")

test_df_SST2 = dataset_test_sampling(df_SST2)

model_df_SST2 = df_for_model(raw_queries_df, test_df_SST2, 'SST2')

SST2_unified = model_df_SST2[["sentence","label","raw query"]]

SST2_unified['label'].replace(0, 'Negative', inplace=True)
SST2_unified['label'].replace(1, 'Positive', inplace=True)

SST2_unified
```

```python
splits = {'train': 'train.csv', 'test': 'test.csv'}
df_COPA = pd.read_csv("hf://datasets/pkavumba/balanced-copa/" + splits["test"])

#df_COPA = pd.read_csv('/content/test__balanced_Copa.csv')

test_df_COPA = dataset_test_sampling(df_COPA)

model_df_COPA = df_for_model(raw_queries_df, test_df_COPA, 'COPA')

#Унификация датасета COPA
COPA_unified = copy.deepcopy(model_df_COPA)
COPA_unified["premise / choice1 / choice2"] = ("Premise: " + COPA_unified["premise"] + " Choice 1: " + COPA_unified["choice1"] + " Choice 2: " + COPA_unified["choice2"])
COPA_unified = COPA_unified[["premise / choice1 / choice2", "label", "raw query"]]

COPA_unified['label'].replace(0, 'Choice 1', inplace=True)
COPA_unified['label'].replace(1, 'Choice 2', inplace=True)

COPA_unified
```

```python
#Classification function

def get_results(text, system_prompt):
    prompt = system_prompt + '\n' + text
    chat = [
        { "role": "user", "content": prompt },
    ]
    prompt = tokenizer.apply_chat_template(chat, tokenize=False, add_generation_prompt=True)

    inputs = tokenizer.encode(prompt, add_special_tokens=False, return_tensors="pt")
    outputs = model.generate(input_ids=inputs.to(model.device), max_new_tokens=500)
    output_decode = tokenizer.decode(outputs[0])

    pattern = r'<start_of_turn>model\s*(.*?)\s*<end_of_turn>'
    match = re.search(pattern, output_decode, re.DOTALL)

    if match:
        answer = match.group(1)
        return answer
    else:
        return "Error" #output_decode
```

```python
SST2_classes = SST2_unified['label'].value_counts().index.tolist()
SST2_classes
```

```python
SST2_classes_str = ', '.join(SST2_classes)

# Function to generate the unification prompt
def generate_unification_prompt(text, classes):
    return f"""
    Provide **exclusively** the category name from the TEXT.
    You can only reply with one of the categories from the list: \"{classes}\".
    Do not provide any additional text or explanation. Do not reply with a category name that is not listed in the \"{classes}\"!

    TEXT:
    \"{text}\"
    """
```

```python
# Function to rerun the unification promt 3 times and then create a separate column with Boolean values for matched and not matched results
def unify_the_output(text, classes):
    result = ""
    class_list = classes.split(', ')  # Split the classes string into a list
    attempt_count = 0  # To keep track of attempts

    while result not in class_list and attempt_count < 3:  # Limit retries to prevent infinite loops
        result = get_results(text, generate_unification_prompt(text, classes)).strip()
        attempt_count += 1  # Increment the attempt counter

    if result not in class_list:
        return result, False  # Return the result and a flag indicating it didn't match
    else:
        return result, True  # Return the result and a flag indicating it matched

# Apply the function to the DataFrame and unpack the tuple into separate columns
SST2_ultimate_results = SST2_df_new.apply(
    lambda row: unify_the_output(row[0], SST2_classes_str),
    axis=1,
    result_type='expand'  # This will ensure that the returned tuple is split into separate columns
)

# Assign the results to appropriate columns
SST2_ultimate_results.columns = ['Classification', 'Matched']

# Optionally, filter out rows where the result did not match
unmatched_rows = SST2_ultimate_results[SST2_ultimate_results['Matched'] == False]
```
